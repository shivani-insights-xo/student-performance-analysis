```python
import pandas as pd

data = {
    'study_time': [1, 2, 3, 4, 5, 2, 3, 4, 1, 5],
    'attendance': [60, 70, 80, 90, 95, 75, 85, 88, 65, 92],
    'marks': [50, 55, 65, 70, 90, 60, 75, 80, 52, 88]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='study_time', y='marks', data=df)
plt.title("Study Time vs Marks")
plt.xlabel("Study Time")
plt.ylabel("Marks")
plt.show()
```

## 📊 Study Time vs Marks

This graph shows the relationship between study time and marks.

💡 Insight: Students who study more tend to score higher marks.


```python
sns.scatterplot(x='attendance', y='marks', data=df)
plt.title("Attendance vs Marks")
plt.xlabel("Attendance")
plt.ylabel("Marks")
plt.show()
```

## 📊 Attendance vs Marks

💡 Insight: Higher attendance leads to better performance.

```python
from sklearn.model_selection import train_test_split

# Features (input)
X = df[['study_time', 'attendance']]

# Target (output)
y = df['marks']

# Split data into training & testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train, X_test
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", error)
```

## 🤖 Machine Learning Model

I used a Linear Regression model to predict student marks based on study time and attendance.

📈 The model was trained on sample data and tested to check its performance.

💡 This shows how machine learning can help predict outcomes using data.
